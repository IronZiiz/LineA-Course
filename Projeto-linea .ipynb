```python
#Seleção de uma região circular com Q3C
#Para selecionar uma região circular, basta informar as coordenadas do centro da seleção e um raio (em graus) nos argumentos da função q3c_radial_query(). 
```

```python
query_2 = "SELECT coadd_object_id, ra ,dec, flags_g, mag_auto_g_dered, mag_auto_r_dered, mag_auto_i_dered, magerr_auto_g, magerr_auto_r, magerr_auto_i FROM des_dr2.main WHERE q3c_radial_query(ra, dec, 15.2, -33.9, 0.7) AND extended_class_coadd < 2 "
query_2    
```

```python
%%time
dados_exemplo_2 = db.fetchall_df(query_2)
```

```python
dados_exemplo_2
```

```python
# Criando o gráfico
plt.plot(dados_exemplo_2["ra"], dados_exemplo_2["dec"], 'k.', alpha=0.1)
plt.xlabel("R.A. (deg)", fontsize=14)
plt.ylabel("Dec. (deg)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
```
